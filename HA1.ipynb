**3. Durch intensive Maßnahmen zur Lärmminderung und -vermeidung könnte der Pegel um durchschnittlich 2 dB gesenkt werden.** 

3.1) _Schätzen_ Sie basierend auf den gegebenen Werten ungefähr ab, wie viele (Werk-)Tage _im Jahr_ es noch immer zu Richtwertüberschreitungen kommen würde und geben Sie das Ergebnis aus.

```python
anzahl_jahr = None # <- diese Variable sollte später die geschätzte Anzahl an Tagen im Jahr mit Richtwert-Überschreitung enthalten
# Hier eigenen Code schreiben ...
LAeq_2dB = {} 

for monat in LAeq:             
    neue_werte = []            

    for j in LAeq[monat]:    
        neuer_wert = j - 2
        neue_werte.append(neuer_wert)  
        LAeq_2dB[monat] = neue_werte

print(LAeq_2dB)

Liste = []
for k in LAeq_2dB:
    anzahl_Tage = 0 
    
    for wert in LAeq_2dB[k]:
        if wert > 60:
            anzahl_Tage += 1

Liste.append(anzahl_Tage)
anzahl_halbjahr = sum(Liste)
anzahl_jahr = Anzahl_halbjahr * 2
print("Anzahl der Richtwertüberschreitungen im Jahr:",anzahl_jahr)



    


```

```python
# Hier ist ein Plausibilitätstest
assert anzahl_jahr > 40
```

3.2) Wie sieht es bei 4 dB durchschnittlicher Pegelverringerung aus?
